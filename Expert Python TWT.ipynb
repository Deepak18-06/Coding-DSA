```python
class Gen:
    def__init__(self,n):
        self.n = n
        self.last = 0
    def __next__(self):
        return self.next
    def next(self):
        if self.last == self.n:
            raise StopIteration()
        rv = self.last**2
        self.last+=1
        return rv
    
```

```python
def solve(arr):
    negative = 0
    positive = 0
    presum = 0
    for i in range(len(arr)):
        presum +=arr[i]
        if arr[i]<0:
            negative+=1
        else:
            positive+=1
    return int(presum/abs(negative-positive))
arr = list(map(int,input().split()))
print(solve(arr))
```
